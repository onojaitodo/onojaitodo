```python
#Used the replace funtion to replace any value thats not a number with a blank space to avoid confusing the program.
def replace(x):
    if type(x) != float:
        return x.replace("","")
```

```python
df.PrimaryColor= df.PrimaryColor.apply(lambda x: replace(x))
```

```python
#Made primary_colors equal do the dataframe portion that showed unique 
primary_colors= df.PrimaryColor.unique()
```

```python
#Create an empty dictionary that would be used to keep track of purchases of colorways. Followed by a for loop that initialized the dictionary to 0.
dictionary={}
for color in primary_colors:
    dictionary[color]= 0
```

```python
#Created a for loop that kept a count and added whenever each color was purchased to the empty dictionary.
for color in primary_colors:
    for i in range(len(df.PrimaryColor)):
        if df.PrimaryColor[i]== color:
            dictionary[color] +=1
```

```python
genders = list(df["Gender"].unique())
```

```python
gender_price_dict={}
gender_count_dict={}
#for gender in the unique list of genders create a key with that gender and value = 0
for gender in genders:
    gender_price_dict[gender]= 0
    gender_count_dict[gender]=0
```

```python
gender_count_dict
```

```python
#This is a function which takes two columns  of a data frame and makes it into a tuple
def pairing(df, col1, col2):
    temp = []
    for x,y in zip(df[col1], df[col2]):
        temp.append((x,y))
    return temp
```

```python
gender_price = []
```

```python
gender_price = pairing(df, "Gender", "Price (INR)")
```

```python
gender_price
```

```python
#for each gender update our dictionaries, 
#price with sum of all prices and count with how many times they occur
for gender in genders:
    for item in gender_price:
        if item[0] == gender:
            gender_price_dict[gender] += item[1]
            gender_count_dict[gender] += 1
    
```

```python
gender_price_dict
```

```python
gender_count_dict
```

```python
#get the average/mean
for gender in genders:
    gender_price_dict[gender] = gender_price_dict[gender]/gender_count_dict[gender]
```

```python
gender_price_dict
```

```python
##Used the keys and values of the gender price dictionary into order to convert them into separate lists.
X = list(gender_price_dict.keys())
Y = list(gender_price_dict.values())
```

```python
##Created an empty lists "temp", then used a for loop to add the list values to the empty list in terms of X and Y.
temp=[]
temp=[]
for x,y in zip(X,Y):
    temp.append((x,y))
```

```python
temp
```

```python
temp_df = pd.DataFrame(temp, columns = ["gender", "Mean"])
```

```python
temp_df
```
